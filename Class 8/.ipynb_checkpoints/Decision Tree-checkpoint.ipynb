```python
import pandas as pd
from pandas.core.dtypes.common import is_numeric_dtype
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
df = pd.read_csv('game data.csv')
```

```python
df
```

```python
df.drop('Days', axis=1, inplace=True)
```

```python
df
```

```python
y = df['Wear Jacket?']
y
```

```python
x = df.drop('Wear Jacket?', axis=1)
x
```

```python
for col in x.columns:
    if is_numeric_dtype(x[col]):
        continue
    else:
        x[col] = LabelEncoder().fit_transform(x[col])
```

```python
x
```

1. Outlook (sunny->1 ; cloudy ->0)
2. Temprature (cold -> 0 ; warm ->1)
3. Routine (Indoor -> 0 ; Outdoor->1)
4. Sunny, Cold , Indoor= (1,0,0) = ??

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(x,y)
```

```python
clf.predict([[1,0,0]]) #  Sunny, Cold , Indoor= ??
```

```python
tree.plot_tree(clf)
```

```python
tree.plot_tree(clf, rounded=True, filled=True, feature_names=x.columns)
```

```python
plt.figure(figsize=(10,10))
tree.plot_tree(clf, rounded=True, filled=True, feature_names=x.columns)
```
